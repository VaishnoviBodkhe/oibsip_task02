```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dataset = pd.read_csv("Unemployment in India.csv")
dataset = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
print(dataset.head())
```

```python
print(dataset.isnull().sum())
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12,10))
sns.heatmap(dataset.corr())
plt.show()
```

```python
dataset.columns=["Region","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Area","longitude","latitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue ="Area", data=dataset)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue ="Area", data=dataset)
plt.show()
```

```python
unemployment = dataset[["Region","Area","Estimated Unemployment Rate"]]
figure = px.sunburst(unemployment, path=["Area","Region"],
                     values="Estimated Unemployment Rate",
                     width=700,height=700,color_continuous_scale="RdY1Gn",
                     title="Unemployment Rate in India")
figure.show()

```
